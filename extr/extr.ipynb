In order to solve E.1 we will first do Exercise 4.4.1 and then estimate the number of distinct elements for the combined hash functions as well.

In E.4.4.1, we are given three hash functions and asked to estimate the number of distinct elements. First, we apply each function to the values from the stream, then convert the results to 5-bit binary, then count the number of trailing zeros for each value, and find  max. After that, we take 2 to the power of that maximum to get the estimate.

By the end, we calculate the final combined estimate taking the mean of the three individual estimates.

```python
stream = [3, 1, 4, 1, 5, 9, 2, 6, 5]

hash_funcs = {
    'h1': lambda x: (2*x+1)%32,
    'h2': lambda x: (3*x+7)%32,
    'h3': lambda x: (4*x)%32
}

def trailing_zeros(n):
    return bin(n)[::-1].find('1')

results = {}
for name, func in hash_funcs.items():
    hashed = [func(x) for x in stream]
    max_zeros = max(trailing_zeros(h) for h in hashed)
    estimate = 2 ** max_zeros
    results[name] = {
        'hashes': hashed,
        'max_trailing_zeros': max_zeros,
        'estimate': estimate
    }

for name, res in results.items():
    print(f"{name}:")
    print(f"  Hash values: {res['hashes']}")
    print(f"  Max trailing zeros: {res['max_trailing_zeros']}")
    print(f"  Estimate: {res['estimate']}\n")

mean_estimate = sum(res['estimate'] for res in results.values()) / 3
print(f"Combined mean estimate: {mean_estimate:.2f}")
```

### Task E.1.a

```python
import random
import numpy as np

stream_8 = [1, 2, 3, 4, 5, 6, 7, 8] * 12 + [1, 2, 3, 4]
stream_3 = [10, 20, 30] * 33 + [10]

def trailing_zeros(n):
    return 5 if n == 0 else bin(n)[::-1].find('1')

hash_params = [(random.randint(1, 31), random.randint(0, 31)) for _ in range(50)]
hash_functions = [lambda x, a=a, b=b: (a * x + b) % 32 for a, b in hash_params]

def run_estimates(stream):
    estimates = []
    for h in hash_functions:
        hashed = [h(x) for x in stream]
        max_zeros = max(trailing_zeros(v) for v in hashed)
        estimates.append(2 ** max_zeros)
    return estimates

estimates_8 = run_estimates(stream_8)
estimates_3 = run_estimates(stream_3)

def combine(estimates):
    mean_val = np.mean(estimates)
    median_val = np.median(estimates)
    grouped = [np.mean(estimates[i:i+10]) for i in range(0, 50, 10)]
    group_median = np.median(grouped)
    return mean_val, median_val, group_median

mean8, med8, groupmed8 = combine(estimates_8)
mean3, med3, groupmed3 = combine(estimates_3)

print("Stream with 8 distinct values:")
print(f"  Mean estimate: {mean8:.2f}")
print(f"  Median estimate: {med8:.2f}")
print(f"  Group-median estimate: {groupmed8:.2f}\n")

print("Stream with 3 distinct values:")
print(f"  Mean estimate: {mean3:.2f}")
print(f"  Median estimate: {med3:.2f}")
```
