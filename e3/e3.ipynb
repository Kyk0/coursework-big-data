#### Task 3.1

Letâ€™s have the following two datasets:
- a) 100 items numbered 1 to 100; 100 baskets, also numbered 1 to 100. Item i is in basket b if
and only if i divides b with no remainder.
Thus, item 1 is in all the baskets, item 2 is in all fifty of the even-numbered baskets, and so
on. For example, basket 12 consists of items {1, 2, 3, 4, 6, 12}, since these are all the integers
that divide 12.
- b) 100 items numbered 1 to 100; 100 baskets, also numbered 1 to 100. Item i is in basket b if
and only if b divides i with no remainder. For example, basket 12 consists of items {12, 24, 36,
48, 60, 72, 84, 96}

```python
def solve_task_3_1a(support_threshold=5, max_item=100, max_basket=100):
    """
    For each of the two datasets described:
      (a) basket b contains all items i that divide b (b % i == 0)
      (b) basket b contains all items i for which b divides i (i % b == 0)

    We compute the support of each item (i.e., in how many baskets it appears)
    and then return the list of items that meet or exceed the given support threshold.
    """

    # 1. Compute support for dataset (a)
    #    Item i is in basket b if i divides b (b % i == 0).
    support_a = {}
    for i in range(1, max_item + 1):
        count_baskets = 0
        for b in range(1, max_basket + 1):
            if b % i == 0:
                count_baskets += 1
        support_a[i] = count_baskets

    # 2. Find frequent items in dataset (a) given the support threshold
    frequent_a = [i for i in range(1, max_item + 1) if support_a[i] >= support_threshold]

    # 3. Compute support for dataset (b)
    #    Item i is in basket b if b divides i (i % b == 0).
    support_b = {}
    for i in range(1, max_item + 1):
        count_baskets = 0
        for b in range(1, max_basket + 1):
            if i % b == 0:
                count_baskets += 1
        support_b[i] = count_baskets

    # 4. Find frequent items in dataset (b) given the support threshold
    frequent_b = [i for i in range(1, max_item + 1) if support_b[i] >= support_threshold]

    return frequent_a, frequent_b

if __name__ == "__main__":
    # We want items with support >= 5
    support_threshold = 5

    frequent_a, frequent_b = solve_task_3_1a(support_threshold=support_threshold)

    print(f"Frequent items in case (a) (support >= {support_threshold}):")
    print(frequent_a)
    print()
    print(f"Frequent items in case (b) (support >= {support_threshold}):")
    print(frequent_b)
```
